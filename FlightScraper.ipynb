```python
import requests
import os
from bs4 import BeautifulSoup
```

```python
url = 'https://www.expedia.com/Flights-Search?flight-type=on&mode=search&trip=oneway&leg1=from%3ASan+Francisco%2C+CA+%28SFO-San+Francisco+Intl.%29%2Cto%3AShanghai+%28PVG-Pudong+Intl.%29%2Cdeparture%3A8%2F26%2F2020TANYT&options=cabinclass%3Aeconomy&passengers=children%3A0%2Cadults%3A1%2Cseniors%3A0%2Cinfantinlap%3AY&fromDate=8%2F26%2F2020&d1=2020-08-26'
```

```python
def extract_url_query(url):
    url_base, params = url.split('?')
    payloads = {}
    for param in params.split('&'):
        name, value = param.split('=')
        payloads[name] = value
    return url_base, payloads
```

```python
url_base, payloads = extract_url_query(url)
```
